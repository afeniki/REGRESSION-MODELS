```python
# Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”].
# The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42
# (for reproducibility). Run a multiple linear regression using the training set and evaluate your
# model on the test set. Answer the following questions:

# What is the Mean Absolute Error (in two decimal places)?
```

```python
scaler = MinMaxScaler()
```

```python
df = df.drop(columns=['date', 'lights'])
```

```python
normalised_data = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
normalised_data
```

```python
features_df = normalised_data.drop(columns = ['Appliances'], axis=1)
```

```python
target_var = normalised_data['Appliances']
```

```python
# split data into training and testing
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target_var, test_size=0.3, random_state = 42)
```

```python
linear_model= LinearRegression()
```

```python
y = linear_model.fit(x_train, y_train)
```

```python
predicted_value = linear_model.predict(x_test)
```

```python
predicted_value
```

# FIND THE MEAN ABSOLUTE ERROR

```python
mae = mean_absolute_error(y_test, predicted_value)
```

```python
round(mae, 2)
```

## What is the Root Mean Squared Error (in three decimal places)?

```python
rmse= np.sqrt(mean_squared_error(y_test, predicted_value))
```

```python
round(rmse, 3)
```

## What is the Residual Sum of Squares (in two decimal places)?

```python
rss = np.sum(np.sqrt(y_test - predicted_value))
```

```python
round(rss, 2)
```

## What is the Coefficient of Determination (in two decimal places)?

```python
r2 = r2_score(y_test, predicted_value)
```

```python
round(r2, 2)
```

## Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
ridge_weight = Ridge(alpha=0.5)
```

```python
ridge_weight.fit(x_train, y_train)
```

```python
lasso_weight = Lasso(alpha = 0.01)
```

```python
lasso_weight.fit(x_train, y_train)
```

```python
def get_weight(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights ).reset_index()
    weights_df.columns = ['features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weight(linear_model, x_train, 'Linear_Mode_weights')
ridge_reg_weights = get_weight(ridge_weight, x_train, 'Ridge_reg_Weights')
lasso_reg_weights = get_weight(lasso_weight, x_train, 'Lasso_reg_weights')
```

```python
final_weights = pd.merge(linear_model_weights, ridge_reg_weights, on='features')
final_weights = pd.merge(final_weights, lasso_reg_weights, on='features')
```

```python
final_weights
```
